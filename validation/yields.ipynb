```python
import ROOT as r
import mplhep as hep
from uproot.pyroot import from_pyroot

import matplotlib.pyplot as plt

hep.style.use(hep.style.CMS)
```

```python
df = r.RDataFrame("Events", "/data/userdata/aaarora/output/run2/data.root")
```

```python
df.Filter("Hbbscore < 0.95 && HighestWjetScore < 0.7", "HScore Cut").Report().Print()
```

```python
hist = df.Histo2D(("h", "2D histogram", 50, 1000, 2000, 50, 0, 600), "ST", "MET", "weight").GetValue()
h = from_pyroot(hist)
```

```python
fig, ax = plt.subplots()
hep.hist2dplot(h, cbar=True, ax=ax)
ax.set_xlabel("ST [GeV]")
ax.set_ylabel("MET [GeV]")
ax.set_title("SIG")
```
